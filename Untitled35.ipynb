```python
from sklearn.base import TransformerMixin

class DataFrameImputer(TransformerMixin):

    def __init__(self):
        """Impute missing values.

        Columns of dtype object are imputed with the most frequent value 
        in column.

        Columns of other types are imputed with mean of column.

        """
    def fit(self, X, y=None):

        self.fill = pd.Series([X[c].value_counts().index[0]
            if X[c].dtype == np.dtype('O') else X[c].mean() for c in X],
            index=X.columns)

        return self

    def transform(self, X, y=None):
        return X.fillna(self.fill)
```

```python
from sklearn.metrics import mean_squared_error 
from sklearn import neighbors
rmse_val = [] #to store rmse values for different k
rmse_val_train = []
for K in range(50):
    K = K+1
    model = neighbors.KNeighborsRegressor(n_neighbors = K)

    model.fit(train_X, train_y)  #fit the model
    
    
    pred = model.predict(test_X) #make prediction on test set
    error = mean_squared_error(test_y, pred) #calculate rmse
    rmse_val.append(error) #store rmse values
    
    pred = model.predict(train_X) #make prediction on test set
    error = mean_squared_error(train_y, pred) #calculate rmse
    rmse_val_train.append(error) #store rmse values
    
    #print('RMSE value for k= ' , K , 'is:', error)

print(rmse_val)
print(rmse_val_train)
```
