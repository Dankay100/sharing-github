```python
from sklearn.feature_extraction.text import CountVectorizer
corpus = df['Description'].dropna()
vectorizer = CountVectorizer(max_features = 300)
X = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names())
```

```python
X.shape
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
words = vectorizer.get_feature_names()
new_corpus = []

for str in corpus:
    w = str.split(" ")
    flag = False
    for w1 in w:
        if w1 in words:
            flag = True
            break
    if flag == True:
        new_corpus.append(str)
        
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(new_corpus)
```

```python
X.shape
```

```python
X.toarray()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
```

```python
kmeans.labels_
```

```python
kmeans.labels_.shape
```

```python
len(new_corpus)
```

```python
result = pd.concat([pd.DataFrame(new_corpus), pd.DataFrame(kmeans.labels_)], axis=1, sort=False)
```

```python
result.shape
```

```python
result.head()
```

```python
symbols = ['Description','ClusterId']
result.columns = symbols
```

```python
result.head()
```

```python
df = pd.merge(df, result, how='left', on='Description')
```

```python
df =  df.dropna(subset=['Description', 'ClusterId'])
```

```python
df.drop_duplicates(inplace = True)
```
